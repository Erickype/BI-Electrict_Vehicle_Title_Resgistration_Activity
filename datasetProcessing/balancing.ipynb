```python
import pandas as pd
```

```python
df = pd.read_csv("./datasets/registrationCleanAlternative.csv")
df.head()
```

```python
label_count = df["label"].value_counts()
print(label_count)
```

```python
label_count_filtered = label_count[label_count >= 1000]

labels_to_keep = label_count_filtered.index.tolist()

df = df[df["label"].isin(labels_to_keep)]

label_count = df["label"].value_counts()
print(label_count)
print(label_count.shape)
```

```python
from imblearn.over_sampling import SMOTENC

X = df.drop('label', axis=1)
y = df['label']
cat_index = [0,1,2,3,4]

oversample = SMOTENC(cat_index, random_state=42)
X_resampled, y_resampled = oversample.fit_resample(X, y)
```

```python
df_balanced = pd.concat([X_resampled, y_resampled], axis='columns')
df_balanced.head()
```

```python
balanced_labels = df_balanced["label"].value_counts()
print(balanced_labels)
```
